```python
from textblob import TextBlob
import codecs
from collections import Counter
```

```python
with codecs.open ("Texts/Dostoyevski_TheGambler.txt", "r",'utf-8') as myfile:
    data=myfile.read().replace('\n', '')
wiki = TextBlob(data)
taggs=wiki.tags
# names=set()
# names= set([str(i[0]) for i in taggs if i[1] == u'NNP' and len(i[0])>1])
name_count=Counter()
for i in taggs:
    if i[1]==u'NNP' and len(i[0])>1:
        name_count[str(i[0])] +=1
#         names.add(str(i))
# 
# names_count.keys()
fop =open('Texts/Dostoyevski_TheGambler_Names.txt','w')
for i,c in name_count.items():
    print i,c
#     fop.write('%s , %s\n' %(i,c))
    fop.write('%s\n' %i)
fop.close()
```

```python
import os
from nltk.tag.stanford import StanfordNERTagger
st =  StanfordNERTagger('classifiers/english.all.3class.distsim.crf.ser.gz', 'stanford-ner.jar')

f1 = open('Dostoyevski_TheGambler.txt','r')
f2 = open('Dostoyevski_TheGambler_Results.txt','w')
f3 = open('Dostoyevski_TheGambler_PERSONS.txt','w')
f4 = open('Dostoyevski_TheGambler_Unique_PERSONS.txt','w')
f5 = open('Dostoyevski_TheGambler_Unique_LOCATIONS.txt','w')

book=f1.read()
persons =[]
#print book
results2= st.tag(book.split())

for name,entity in results2:
    print name +" " + entity
    f2.write(name +" " + entity+"\n" )
#     if entity == "PERSON":
#         f3.write(name +"\n")
#         if name not in persons:
#             persons.append(name)
# for k in persons:
#     f4.write(k+"\n")

# Words - Tags [Tags = Person, Location, Organization, O(ther)]
```

```python
for name,entity in results2:
    if entity == "PERSON":
        print name
        f3.write(name +"\n")

# Persons
```

```python
for name,entity in results2:
    if entity == "PERSON":
        if name not in persons:
            persons.append(name)
persons1=[]
for elem in persons:
    persons1.append(elem.strip('.').strip('"').strip(','))
for k in persons1:
    if len(k) == 2:
        persons1.remove(k)
persons1.sort()
print persons1
# for k in persons1:
#     print k
#     f4.write(k+"\n")
    
# Unique Persons
```

```python
locations=[]
for name,entity in results2:
    if entity == "LOCATION":
        if name not in locations:
            locations.append(name)
locations.sort()
for k in locations:
    print k
    f5.write(k+"\n")
    
# (Unique) Locations
```
